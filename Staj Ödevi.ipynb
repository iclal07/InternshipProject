```python
features=data.columns
def outliers(df, features):#aykırı değere bakmak için kullanılır
    """Count the number of outliers for each feature using the IQR"""
    num_outliers = []
    pct_outliers = []
    total_rows = []
    interquartile_range = []
    for i in features:
        Q1 = df[i].quantile(.25)
        Q3 = df[i].quantile(.75)
        IQR = Q3 - Q1
        outliers = len(df[(df[i] < (Q1-1.5*IQR)) | (df[i] > (Q3+1.5*IQR))])
        rows = len(df[i])
        pct = outliers/rows
        interquartile_range.append(IQR)
        num_outliers.append(outliers)
        pct_outliers.append(pct)
        total_rows.append(rows)
        
    count_outliers = pd.DataFrame({'Feature': features
                               , 'Num_Outliers': num_outliers
                                , 'Percent_Outliers': pct_outliers
                                , 'IQR': interquartile_range
                               , 'Total_Rows': total_rows}).sort_values('Percent_Outliers', ascending=False)
    return count_outliers

count_outliers = outliers(df=data, features=features)
count_outliers
```

```python
#elbow metodu k sayısını verir. K sayısı en optimum değerini bulmamızı sağlar
```

```python
wcss=[]

for i in range(1,20):
    k_means=KMeans(n_clusters=i,init="k-means++")
    k_means.fit(df)
    wcss.append(k_means.inertia_)
plt.plot(range(1,20),wcss,"o")
plt.title("Elbow Methodu")
plt.xlabel("Küme sayısı") #k sayısının ağırlığını belirler. Kaç tane k olursa o kadar kümem olur
plt.ylabel("wcss")
plt.show()    
    
```
